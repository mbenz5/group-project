```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
location = 'nyc-rolling-sales.csv'
df=pd.read_csv(location)
```

```python
df.head()
```

```python
sns.lmplot(x='RESIDENTIAL UNITS',y='YEAR BUILT',data=df)
```

```python
plt.subplots(figsize=(18,10))
sns.boxplot(data=df)
```

```python
stats_df=df.drop(['Unnamed: 0','BOROUGH','NEIGHBORHOOD'],axis=1)
stats_df.head()
```

```python
plt.subplots(figsize=(18,10))
sns.boxplot(data=stats_df)
```

```python
bins=[10000,11000,12000]
stats_df['ZIP CODE'] = pd.cut(stats_df['ZIP CODE'], bins)
df.head()
```

```python
df.loc[(df['ZIP CODE'] >= 10000) & (df['ZIP CODE'] < 12000)]
plt.subplots(figsize=(18,10))
sns.violinplot(x='YEAR BUILT',y='ZIP CODE', data=stats_df),
```

```python
sns.distplot(df['ZIP CODE'])
```
